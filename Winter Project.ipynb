```python
import cv2
import numpy as np
import math as m
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def nothing(x):
    print(x)
```

```python
src_img = cv2.imread('bolt2.jpg',1)
gauss = cv2.GaussianBlur(src_img,(5,5),6)
gray = cv2.cvtColor(gauss,cv2.COLOR_BGR2GRAY)
ret,thres = cv2.threshold(gray,127,255,cv2.THRESH_TOZERO_INV)
#thres = cv2.cvtColor(thres,cv2.COLOR_B)
cv2.imshow('thresh',thres)
cv2.waitKey(0)
cv2.namedWindow('Edge Detect')
cv2.createTrackbar('Thresh1','Edge Detect',0,255,nothing)
cv2.createTrackbar('Thresh2','Edge Detect',0,255,nothing)
t1 = t2 = 0
while(1):
    canny = cv2.Canny(thres,t1,t2)
    cv2.imshow('Edge Detect',canny)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:#Esc Key for terminating
        break
    t1 = cv2.getTrackbarPos('Thresh1','Edge Detect')
    t2 = cv2.getTrackbarPos('Thresh2','Edge Detect')
cv2.destroyAllWindows()   
```

```python
#ret,canny = cv2.threshold(canny,127,255,0) 
cont, hier = cv2.findContours(canny,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE) 
```

```python
img2 = img[:,800:1300]
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
cv2.imshow('cropped',img2)
cv2.waitKey(0)
```

```python
for r in range(img2.shape[0]):
    if 255 in img2[r,:]:
        min1 = r
        break
for r in range(img2.shape[0]-1,-1,-1):
    if 255 in img2[r,:]:
        max1 = r
        break
print(min1)
print(max1)
mid = (min1+max1)/2
print(mid)
```

```python
img2[int(mid),:] = 255
cv2.imshow('midline',img2)
cv2.waitKey(0)
```

```python
upper = []
mid = int(mid)
for col in range(img2.shape[1]):
    for row in range(mid):
        if int(img2[row,col]) == 255:
            tup = (row,col)
            upper.append(tup)
            tup = ()

print(upper)            
```

```python
lower = []
for col in range(img2.shape[1]):
    for row in range(mid+1,img2.shape[0]):
        if int(img2[row,col]) == 255:
            tup = (row,col)
            lower.append(tup)
            tup = ()

print(lower)   
```

```python
cv2.imshow('image',img2)
cv2.waitKey(0)
```

```python
direction = []
for x in range(len(upper)-1):
    if upper[x][0] > upper[x+1][0]:
        tup = (1,upper[x][0],upper[x][1])
        direction.append(tup)
        tup = ()
    elif upper[x][0] < upper[x+1][0]:
        tup = (0,upper[x][0],upper[x][1])
        direction.append(tup)
        tup = ()
    else:
        continue

print(direction)
```

```python
crest = root = []
for x in range(len(direction)-1):
    if direction[x][0] == 1 and direction[x+1][0] == 0:
        tup = (direction[x][1],direction[x][2])
        crest.append(tup)
        tup = ()
    elif direction[x][0] == 0 and direction[x+1][0] == 1:
        tup = (direction[x][1],direction[x][2])
        root.append(tup)
        tup = ()
```

```python
img3 = img2
for coord in crest:
    cv2.circle(img3,(coord[1],coord[0]),5,155,-1)
cv2.imshow('crestpoints',img3)
cv2.waitKey(0)
```
